```python
import cv2
import numpy as np

def wiener_filter(img, kernel_size=5):
  kernel = np.ones((kernel_size, kernel_size)) / (kernel_size**2)
  img_fft = np.fft.fft2(img)
  kernel_fft = np.fft.fft2(kernel, s=img.shape)
  kernel_fft_conj = np.conj(kernel_fft)
  wiener = kernel_fft_conj / (np.abs(kernel_fft)**2 + 0.01)

  result_fft = wiener * img_fft
  result = np.abs(np.fft.ifft2(result_fft))

  return result
```

```python
img = cv2.imread(os.path.join(blur_path, sample), 0)
deblurred = wiener_filter(img)

plt.imshow(deblurred, cmap='gray')
plt.title("Deblurred Image")
plt.show()
```

```python
from skimage.metrics import peak_signal_noise_ratio, structural_similarity
import numpy as np

clean_gray = cv2.imread(os.path.join(clean_path, sample), 0)

clean_gray = clean_gray.astype(np.float32)
deblurred = deblurred.astype(np.float32)
deblurred = np.clip(deblurred, 0, 255)
psnr = peak_signal_noise_ratio(clean_gray, deblurred, data_range=255)
ssim = structural_similarity(clean_gray, deblurred, data_range=255)
print("PSNR: ", psnr)
print("SSIM: ", ssim)
```

```python
import random
common_list = sorted(list(common))
subset = random.sample(common_list, 30)
print("Number of images selected: ", len(subset))
print(subset[:5])
```

```python
def deblur_image_gray(image_path):
  img = cv2.imread(image_path, 0)
  deblurred = wiener_filter(img)
  deblurrd = np.clip(deblurred, 0, 255).astype(np.float32)
  return img.astype(np.float32), deblurred
```

```python
def get_detection_stats(result):
  boxes = result.boxes
  if boxes is None or len(boxes) == 0:
    return 0,0.0
  confs = boxes.conf.cpu().numpy()
  return len(confs), float(np.mean(confs))
```

```python
import pandas as pd
rows = []
for fname in subset:
  blur_file = os.path.join(blur_path, fname)
  clean_file = os.path.join(clean_path, fname)

  blur_gray, deblurred = deblur_image_gray(blur_file)
  clean_gray = cv2.imread(clean_file, 0).astype(np.float32)

  psnr = peak_signal_noise_ratio(clean_gray, deblurred, data_range=255)
  ssim = structural_similarity(clean_gray, deblurred, data_range=255)

  tmp_file = os.path.join("/content", f"tmp_{fname}")
  cv2.imwrite(tmp_file, deblurred.astype(np.uint8))

  res_blur = model(blur_file, verbose=False)[0]
  res_deblur = model(tmp_file, verbose=False)[0]
  res_clean = model(clean_file, verbose=False)[0]

  n_blur, conf_blur = get_detection_stats(res_blur)
  n_deblur, conf_deblur = get_detection_stats(res_deblur)
  n_clean, conf_clean = get_detection_stats(res_clean)

  rows.append({
    "file": fname,
    "psnr": psnr,
    "ssim": ssim,
    "blur_detections": n_blur,
    "deblur_detections": n_deblur,
    "clean_detections": n_clean,
    "blur_conf": conf_blur,
    "deblur_conf": conf_deblur,
    "clean_conf": conf_clean
})
df = pd.DataFrame(rows)
df.head()
```

```python
summary = df[[
    "psnr",
    "ssim",
    "blur_detections",
    "deblur_detections",
    "clean_detections",
    "blur_conf",
    "deblur_conf",
    "clean_conf"
]].mean()
print(summary)
```

```python
df["det_change"] = df["deblur_detections"] - df["blur_detections"]
df["conf_change"] = df["deblur_conf"] - df["blur_conf"]

print("Images with more detections after deblurring: ", (df["det_change"] > 0).sum())
print("Images with same detections: ", (df["det_change"] == 0).sum())
print("Images with fewer detections: ", (df["det_change"] < 0).sum())

print("Average detection change: ", df["det_change"].mean())
print("Average confidence change: ", df["conf_change"].mean())
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["det_change"], bins=10)
plt.title("Change in Number of Detections After Deblurring")
plt.xlabel("Deblurred detections - Blurred detections")
plt.ylabel("Number of images")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["conf_change"], bins=10)
plt.title("Change in Average Confidence After Deblurring")
plt.xlabel("Deblurred confidence - Blurred confidence")
plt.ylabel("Number of images")
plt.show()
```
